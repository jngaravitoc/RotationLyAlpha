```python
%pylab inline
font = {'size':25, 'family':'serif'}
matplotlib.rc('font', **font)
```

```python
def loading_data(filename):
    data = loadtxt("../data/Central/" + filename)
    kz = data[:,5]
    x = data[:,6]
    nscatt = data[:,8]
    return kz, x, nscatt
```

```python
kz5t0, x5t0, nscatt = loading_data("Rotating0Sphere5t.ascii")
kz5t100, x5t100, nscatt = loading_data("Rotating100Sphere5t.ascii")
kz5t200, x5t200, nscatt = loading_data("Rotating200Sphere5t.ascii")
kz5t300, x5t300, nscatt = loading_data("Rotating300Sphere5t.ascii")

kz6t0, x6t0, nscatt = loading_data("Rotating0Sphere6t.ascii")
kz6t100, x6t100, nscatt = loading_data("Rotating100Sphere6t.ascii")
kz6t200, x6t200, nscatt = loading_data("Rotating200Sphere6t.ascii")
kz6t300, x6t300, nscatt = loading_data("Rotating300Sphere6t.ascii")

kz7t0, x7t0, nscatt = loading_data("Rotating0Sphere7t.ascii")
kz7t100, x7t100, nscatt = loading_data("Rotating100Sphere7t.ascii")
kz7t200, x7t200, nscatt = loading_data("Rotating200Sphere7t.ascii")
kz7t300, x7t300, nscatt = loading_data("Rotating300Sphere7t.ascii")
```

```python
x5t0, y5t0, z = hist(kz5t0, bins=20)#,  normed=True)
x5t100, y5t100, z = hist(kz5t100, bins=20)#, normed=True)
x5t200, y5t200, z = hist(kz5t200, bins=20)#,  normed=True)
x5t300, y5t300, z = hist(kz5t300, bins=20)#,  normed=True)

x6t0, y6t0, z = hist(kz6t0, bins=20)#,  normed=True)
x6t100, y6t100, z = hist(kz6t100, bins=20)#,  normed=True)
x6t200, y6t200, z = hist(kz6t200, bins=20)#,  normed=True)
x6t300, y6t300, z = hist(kz6t300, bins=20)#, normed=True)

x7t0, y7t0, z = hist(kz7t0, bins=20)#, normed=True)
x7t100, y7t100, z = hist(kz7t100, bins=20)#, normed=True)
x7t200, y7t200, z = hist(kz7t200, bins=20)#, normed=True)
x7t300, y7t300, z = hist(kz7t300, bins=20)#, normed=True)
```

```python
figure(figsize=(9.5, 9.5))
plot(y5t0[:-1],x5t0[:]*20.0/(size(kz5t0)), label=r"$\rm{\tau = 10^5, V=0}$")
plot(y5t100[:-1],x5t100[:]*20.0/(size(kz5t100)), label=r"$\rm{\tau = 10^5, V=100}$")
plot(y5t200[:-1],x5t200[:]*20.0/(size(kz5t200)), label=r"$\rm{\tau = 10^5, V=200}$")
plot(y5t300[:-1],x5t300[:]*20.0/(size(kz5t300)), label=r"$\rm{\tau = 10^5, V=300}$")

plot(y6t0[:-1],x6t0[:]*20.0/(size(kz6t0)), label=r"$\rm{\tau = 10^6, V=0}$")
plot(y6t100[:-1],x6t100[:]*20.0/(size(kz6t100)), label=r"$\rm{\tau = 10^6, V=100}$")
plot(y6t200[:-1],x6t200[:]*20.0/(size(kz6t200)), label=r"$\rm{\tau = 10^6, V=200}$")
plot(y6t300[:-1],x6t300[:]*20.0/(size(kz6t300)), label=r"$\rm{\tau = 10^6, V=300}$")

plot(y7t0[:-1],x7t0[:]*20.0/(size(kz7t0)), label=r"$\rm{\tau = 10^7, V=0}$")
plot(y7t100[:-1],x7t100[:]*20.0/(size(kz7t100)), label=r"$\rm{\tau = 10^7, V=100}$")
plot(y7t200[:-1],x7t200[:]*20.0/(size(kz7t200)), label=r"$\rm{\tau = 10^7, V=200}$")
plot(y7t300[:-1],x7t300[:]*20.0/(size(kz7t300)), label=r"$\rm{\tau = 10^7, V=300}$")

ylim(0.85, 1.08)
legend(loc="best", prop={'size':18}, ncol = 2)
xlabel(r"$\rm{\mu}$", fontsize=30)
ylabel(r"$\rm{F(\mu)}$", fontsize=30)
savefig("../paper/muhisto.eps",bbox_inches='tight', format='eps', transparent=True)
```
