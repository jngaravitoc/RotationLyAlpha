```python
def loading_data(filename):
    data = loadtxt("../data/Central/" + filename)
    kz = data[:,5]
    x = data[:,6]
    nscatt = data[:,8]
    return kz, x, nscatt
```

```python
kz, x, nscatt = loading_data("Rotating300Sphere6t.ascii")
```

```python
histo = hist2d(kz, log10(nscatt), bins=(30 , 30))
colorbar()
```

```python
figure(figsize=(10.5, 9))
histo = hist2d(x, abs(kz), bins=(30 , 30))
xlabel("$x$", fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=18)
ylabel(r"$\mathrm{|\cos( \theta)|}$", fontsize=25)
colorbar()
savefig("muvsx.png")
```
