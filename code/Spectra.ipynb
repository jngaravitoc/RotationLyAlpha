```python
## Plot data Homogeneous Model 

x = flarray0HOM[:,0]
y = flarray0HOM[:,1]
x1 = flarray1HOM[:,0]
y1 = flarray1HOM[:,1]
x2 = flarray2HOM[:,0]
y2 = flarray2HOM[:,1]
x3 = flarray3HOM[:,0]
y3 = flarray3HOM[:,1]

V = (- 12.8451673301 * x)

figure(num=None, figsize=(11.0,9.5))
ax1 = subplot(1, 1, 1)
ax1.plot(x, y, label=r'$\mathrm{50 km}$ $\mathrm{s^{-1}}$', linewidth = 2.5, ls = '-.', c = 'k')
#plt.plot(V, y, label=r'$\mathrm{50 km}$ $\mathrm{s^{-1}}$', linewidth = 2.5)
ax1.plot(x1,y1, label='$\mathrm{100 km}$ $\mathrm{s^{-1}}$', linewidth = 2.5, ls = ':', c = 'k')
ax1.plot(x2,y2, label='$\mathrm{200 km}$ $\mathrm{s^{-1}}$', linewidth = 2.5, ls = '--', c = 'k')
ax1.plot(x3,y3, label='$\mathrm{300 km}$ $\mathrm{s^{-1}}$', linewidth = 2.5, ls = '-', c = 'k')
ax1.legend(loc='upper left', fontsize = 20)
ax1.set_xlabel (r'$\mathrm{x}$', fontsize=35)
ax1.set_ylabel ('$\mathrm{Intensity}$ $\mathrm{(Arbitrary Units)}$', fontsize=35)
ax1.tick_params(axis='both', which='major', labelsize=18)

new_tick_locations = np.array([0, .166, .3333, .5, .666, .833, 1])

ax2 = ax1.twiny()
ax2.set_xticks(new_tick_locations)
ax2.tick_params(labelsize=18)
ax2.set_xticklabels(['-768','-512','-256', '0', '256', '512', '768'])
ax2.set_xlabel(r"$\mathrm{Homogeneous}$ ($\mathrm{\tau_{H}=10^7}$)", fontsize=35)

savefig('SpectraDifVelocitiesHOM.png')

plt.show()
print  (1/6.0)*5
print 12.8*60, 12.8*40, 12.8*20
```
