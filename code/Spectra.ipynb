# Measuring FWHM

```python
def FWHMHomogeneous(x, y, name, mode):
    h = amax(y)
    half = h/2.0
    Y = []
    X = []
    A = []
    if(mode == 7):
        m = 15
    elif(mode==6):
        m = 8
    elif(mode==5):
        m = 4.0
    for i in range(len(x)):
        if(np.abs(x[i])>m):
            X.append(x[i])
            Y.append(y[i])
    for i in range(len(X)):
        a = np.abs(Y[i]-half)
        A.append(a)
    z = amin(A)
    xx = A.index(z)
    FWHM = np.abs(2*X[xx])
    f = open('../data/HomogeneousWidhts'+name+'.dat', 'w')
    f.write(str(FWHM))
    f.close
    #print FWHM
    
def FWHMcentral(x, y, name, mode):
    h = amax(y)
    half = h/2.0
    #print half
    Y = []
    X = []
    A = []
    if(mode == 7):
        m = 15
    elif(mode==6):
        m = 10
    elif(mode==5):
        m = 4
    for i in range(len(x)):
        if(np.abs(x[i])>m):
            X.append(x[i])
            Y.append(y[i])
    for i in range(len(X)):
        a = np.abs(Y[i]-half)
        A.append(a)
    z = amin(A)
    xx = A.index(z)
    FWHM = np.abs(2*X[xx])
    f = open('../data/CentralWidhts'+name+'.dat', 'w')
    f.write(str(FWHM))
    f.close
    #print FWHM
    
    
    
FWHMHomogeneous(x70h, y70h, "t7v0hom", 7)
FWHMHomogeneous(x71h, y71h, "t7v100hom", 7)
FWHMHomogeneous(x72h, y72h, "t7v200hom", 7)
FWHMHomogeneous(x73h, y73h, "t7v300hom", 7)
FWHMHomogeneous(x60h, y60h, "t6v0hom", 6)
FWHMHomogeneous(x61h, y61h, "t6v100hom", 6)
FWHMHomogeneous(x62h, y62h, "t6v200hom", 6)
FWHMHomogeneous(x63h, y63h, "t6v300hom", 6)
FWHMHomogeneous(x50h, y50h, "t5v0hom", 5)
FWHMHomogeneous(x51h, y51h, "t5v100hom", 5)
FWHMHomogeneous(x52h, y52h, "t5v200hom", 5)
FWHMHomogeneous(x53h, y53h, "t5v300hom", 5)

FWHMcentral(x70, y70, "t7v0", 7)
FWHMcentral(x71, y71, "t7v100", 7)
FWHMcentral(x72, y72, "t7v200", 7)
FWHMcentral(x73, y73, "t7v300", 7)
FWHMcentral(x60, y60, "t6v0", 6)
FWHMcentral(x61, y61, "t6v100", 6)
FWHMcentral(x62, y62, "t6v200", 6)
FWHMcentral(x63, y63, "t6v300", 6)
FWHMcentral(x50, y50, "t5v0", 5)
FWHMcentral(x51, y51, "t5v100", 5)
FWHMcentral(x52, y52, "t5v200", 5)
FWHMcentral(x53, y53, "t5v300", 5)

#X.index(35.5)
#print X
#list.index?
```
