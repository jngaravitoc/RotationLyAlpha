```python
from numpy import *
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
flarray0HOM = np.loadtxt("../data/Homogeneous_Histograms/HistogramRotating0Sphere7tHOM.ascii")
flarray1HOM = np.loadtxt("../data/Homogeneous_Histograms/HistogramRotating100Sphere7tHOM.ascii")
flarray2HOM = np.loadtxt("../data/Homogeneous_Histograms/HistogramRotating200Sphere7tHOM.ascii")
flarray3HOM = np.loadtxt("../data/Homogeneous_Histograms/HistogramRotating300Sphere7tHOM.ascii")

flarray0 = np.loadtxt("../data/Central_Histograms/HistogramRotating0Sphere7t.ascii")
flarray1 = np.loadtxt("../data/Central_Histograms/HistogramRotating100Sphere7t.ascii")
flarray2 = np.loadtxt("../data/Central_Histograms/HistogramRotating200Sphere7t.ascii")
flarray3 = np.loadtxt("../data/Central_Histograms/HistogramRotating300Sphere7t.ascii")
```

```python
## Plot data Homogeneous Model 

x = flarray0HOM[:,0]
y = flarray0HOM[:,1]
x1 = flarray1HOM[:,0]
y1 = flarray1HOM[:,1]
x2 = flarray2HOM[:,0]
y2 = flarray2HOM[:,1]
x3 = flarray3HOM[:,0]
y3 = flarray3HOM[:,1]

figure(num=None, figsize=(11.0,9.5))

l0 = plt.plot(x, y, label=r'$\mathrm{50 km}$ $\mathrm{s^{-1}}$', linewidth = 2.5)
l1 = plt.plot(x1,y1, label='$\mathrm{100 km}$ $\mathrm{s^{-1}}$', linewidth = 2.5)
l2 = plt.plot(x2,y2, label='$\mathrm{200 km}$ $\mathrm{s^{-1}}$', linewidth = 2.5)
l3 = plt.plot(x3,y3, label='$\mathrm{300 km}$ $\mathrm{s^{-1}}$', linewidth = 2.5)
ll = plt.legend(loc='upper left', fontsize = 20)
plt.xlabel (r'$\mathrm{x}$', fontsize=35)
plt.ylabel ('$\mathrm{Intensity}$ $\mathrm{(Arbitrary Units)}$', fontsize=35)
plt.title(r"$\mathrm{Homogeneous}$ , $\mathrm{\tau_{H}=10^7}$", fontsize=35)
plt.tick_params(axis='both', which='major', labelsize=18)

savefig('SpectraDifVelocitiesHOM.png')

plt.show()
```
