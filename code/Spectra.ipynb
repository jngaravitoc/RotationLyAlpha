# Measuring FWHM

```python
def FWHMHomogeneous(x, y, name, mode):
    h = amax(y)
    half = h/2.0
    Y = []
    X = []
    A = []
    if(mode == 7):
        m = 15
    elif(mode==6):
        m = 8
    elif(mode==5):
        m = 4.0
    for i in range(len(x)):
        if(np.abs(x[i])>m):
            X.append(x[i])
            Y.append(y[i])
    for i in range(len(X)):
        a = np.abs(Y[i]-half)
        A.append(a)
    z = amin(A)
    xx = A.index(z)
    FWHM = np.abs(2*X[xx])
    f = open('../data/HomogeneousWidhts'+name+'.dat', 'w')
    f.write(str(FWHM))
    f.close
    #print FWHM
    
def FWHMcentral(x, y, name, mode):
    h = amax(y)
    half = h/2.0
    #print half
    Y = []
    X = []
    A = []
    if(mode == 7):
        m = 15
    elif(mode==6):
        m = 10
    elif(mode==5):
        m = 4
    for i in range(len(x)):
        if(np.abs(x[i])>m):
            X.append(x[i])
            Y.append(y[i])
    for i in range(len(X)):
        a = np.abs(Y[i]-half)
        A.append(a)
    z = amin(A)
    xx = A.index(z)
    FWHM = np.abs(2*X[xx])
    f = open('../data/CentralWidhts'+name+'.dat', 'w')
    f.write(str(FWHM))
    f.close
    #print FWHM
    
    
    
FWHMHomogeneous(x70h, y70h, "t7v0hom", 7)
FWHMHomogeneous(x71h, y71h, "t7v100hom", 7)
FWHMHomogeneous(x72h, y72h, "t7v200hom", 7)
FWHMHomogeneous(x73h, y73h, "t7v300hom", 7)
FWHMHomogeneous(x60h, y60h, "t6v0hom", 6)
FWHMHomogeneous(x61h, y61h, "t6v100hom", 6)
FWHMHomogeneous(x62h, y62h, "t6v200hom", 6)
FWHMHomogeneous(x63h, y63h, "t6v300hom", 6)
FWHMHomogeneous(x50h, y50h, "t5v0hom", 5)
FWHMHomogeneous(x51h, y51h, "t5v100hom", 5)
FWHMHomogeneous(x52h, y52h, "t5v200hom", 5)
FWHMHomogeneous(x53h, y53h, "t5v300hom", 5)

FWHMcentral(x70, y70, "t7v0", 7)
FWHMcentral(x71, y71, "t7v100", 7)
FWHMcentral(x72, y72, "t7v200", 7)
FWHMcentral(x73, y73, "t7v300", 7)
FWHMcentral(x60, y60, "t6v0", 6)
FWHMcentral(x61, y61, "t6v100", 6)
FWHMcentral(x62, y62, "t6v200", 6)
FWHMcentral(x63, y63, "t6v300", 6)
FWHMcentral(x50, y50, "t5v0", 5)
FWHMcentral(x51, y51, "t5v100", 5)
FWHMcentral(x52, y52, "t5v200", 5)
FWHMcentral(x53, y53, "t5v300", 5)

#X.index(35.5)
#print X
#list.index?
```

```python
fig = figure(num=None, figsize=(19,7))
fig.subplots_adjust(hspace=0.00, wspace=0.0,
                    top=0.98, right=0.95, bottom=0.05)
#plt.tick_params(axis='both', which='major', labelsize=25)

v = [0, 100, 200, 300]

ax = fig.add_subplot(121)
p1, = ax.plot(v/(w70h*12.85), (w7h*12.85)/(w70h*12.85),  linewidth = 2.5, marker = '^', ms=10, c= 'k', linestyle = '--')
p2, = ax.plot(v/(w60h*12.85), (w6h*12.85)/(w60h*12.85),  linewidth = 2.5, marker = 's', ms=10, c= 'k', linestyle = '--')
p3, = ax.plot(v/(w50h*12.85), (w5h*12.85)/(w50h*12.85),  linewidth = 2.5, marker = 'o', ms=10, c= 'k', linestyle = '--')
ax.set_xlim([0, 2.7])
ax.legend([p1, p2, p3], [r'$\tau=10^{7}$',r'$\tau=10^{6}$', r'$\tau=10^{5}$'], loc ='upper left', ncol=1, fontsize=20)

ax = fig.add_subplot(122)
p1, = ax.plot(v/w50, (w5*12.85)/(w50*12.85), linewidth = 2.5, marker = 'o', ms=10, c= 'k')
p2, = ax.plot(v/w60, (w6*12.85)/(w60*12.85), linewidth = 2.5, marker = 's', ms=10, c= 'k')
p3, = ax.plot(v/w70, (w7*12.85)/(w70*12.85), linewidth = 2.5, marker = '^', ms=10, c= 'k')
ax.set_ylim([1.0, 2.2])
ax.legend([p1, p2, p3], [r'$\tau=10^{5}$',r'$\tau=10^{6}$', r'$\tau=10^{7}$'], loc ='upper left', ncol=1, fontsize=20)
ax.yaxis.set_major_formatter(plt.NullFormatter())
#Axes limits 

#plt.xlim([-10, 310])
#plt.ylim([100, 735])

# Legend 



ax.text(-5.0, 0.85, r'$\mathrm{V_{max}/FWHM_o}$', fontsize=35)
ax.text (-34, 1.8, r'$\mathrm{FWHM/FWHM_o}$', fontsize=35, rotation='vertical')


savefig('../paper/f2.eps', format='eps', transparent=True, bbox_inches='tight')
```
