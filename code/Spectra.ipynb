```python
from numpy import *
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
flarray0HOM = np.loadtxt("../data/Homogeneous_Histograms/HistogramRotating0Sphere7tHOM.ascii")
flarray1HOM = np.loadtxt("../data/Homogeneous_Histograms/HistogramRotating100Sphere7tHOM.ascii")
flarray2HOM = np.loadtxt("../data/Homogeneous_Histograms/HistogramRotating200Sphere7tHOM.ascii")
flarray3HOM = np.loadtxt("../data/Homogeneous_Histograms/HistogramRotating300Sphere7tHOM.ascii")

flarray0 = np.loadtxt("../data/Central_Histograms/HistogramRotating0Sphere7t.ascii")
flarray1 = np.loadtxt("../data/Central_Histograms/HistogramRotating100Sphere7t.ascii")
flarray2 = np.loadtxt("../data/Central_Histograms/HistogramRotating200Sphere7t.ascii")
flarray3 = np.loadtxt("../data/Central_Histograms/HistogramRotating300Sphere7t.ascii")
```

```python
## Plot data Homogeneous Model 

x = flarray0HOM[:,0]
y = flarray0HOM[:,1]
x1 = flarray1HOM[:,0]
y1 = flarray1HOM[:,1]
x2 = flarray2HOM[:,0]
y2 = flarray2HOM[:,1]
x3 = flarray3HOM[:,0]
y3 = flarray3HOM[:,1]

l0 = plt.plot(x, y, label=r'$0 Kms^{-1}$')
l1 = plt.plot(x1,y1, label='$100 Kms^{-1}$')
l2 = plt.plot(x2,y2, label='$200 Kms^{-1}$')
l3 = plt.plot(x3,y3, label='$300 kms^{-1}$')
ll = plt.legend(loc='upper left')
plt.xlabel (r'$x$', fontsize=20)
plt.ylabel ('Intensity (Arbitrary units)', fontsize=20)
plt.title(r"Homogeneous , $\tau_{H}=10^7$", fontsize=20)

savefig('SpectraDifVelocitiesHOM.png')

plt.show()
```

```python
## Plot data Cetral Model 

x = flarray0[:,0]
y = flarray0[:,1]
x1 = flarray1[:,0]
y1 = flarray1[:,1]
x2 = flarray2[:,0]
y2 = flarray2[:,1]
x3 = flarray3[:,0]
y3 = flarray3[:,1]

l0 = plt.plot(x, y, label=r'$0 Kms^{-1}$')
l1 = plt.plot(x1,y1, label='$100 Kms^{-1}$')
l2 = plt.plot(x2,y2, label='$200 Kms^{-1}$')
l3 = plt.plot(x3,y3, label='$300 kms^{-1}$')
ll = plt.legend(loc='upper left')
plt.xlabel (r'$x$', fontsize=20)
plt.ylabel ('Intensity (Arbitrary units)', fontsize=20)
plt.title(r"Central , $\tau_{H}=10^7$", fontsize=20)

savefig('SpectraDifVelocitiesCentral.png')

plt.show()
```
