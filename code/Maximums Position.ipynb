```python
#Final Plot----- Homogoneous and Central vs Vmax

#plot maximums position versus velocity for each tau, Homogeneous Distribution and observer at 0 theta
#refinar estos datos despues con el metodo que propuso Jaime
figure(num=None, figsize=(9.5,9))
xcentral5 = np.array([3.56, 3.4, 3.36, 3.6])
xcentral6 = np.array([6.72,6.9,7.0,6.7])
xcentral7 = np.array([14.62,14.76,14.83,15.1])

xhom5 = np.array([3.0,2.8,0.0, 0.0, 0.0])
xhom6 = np.array([4.8, 5.2, 5.24, 0.0, 0.0])
xhom7 = np.array([10.5, 9.9, 10.5, 11.5, 0.0])

xVcentral5 = xcentral5*12.85
xVcentral6 = xcentral6*12.85
xVcentral7 = xcentral7*12.85

xVhom5 = xhom5*12.85
xVhom6 = xhom6*12.85
xVhom7 = xhom7*12.85

y = [0, 100, 200, 300]
yhom = [0, 50, 100, 200, 300]

p5hom, = plot(yhom,xVhom5,'-r*', linewidth = 3.5)
p6hom, = plot(yhom,xVhom6,'-b^', linewidth = 3.5)
p7hom, = plot(yhom,xVhom7,'-go', linewidth = 3.5)

p5, = plot(y,xVcentral5,'--r*', linewidth = 3.5)
p6, = plot(y,xVcentral6,'--b^', linewidth = 3.5)
p7, = plot(y,xVcentral7,'--go', linewidth = 3.5)

legend([p5hom, p6hom, p7hom], [r"$\tau_{H}=10^{5}$",r"$\tau_{H}=10^6$",r"$\tau_{H}=10^7$"], loc = 'right',bbox_to_anchor=(1.0, 0.9),
          ncol=3, fontsize =20)

plt.xlabel (r'$\mathrm{V_{max}}$($\mathrm{km}$ $\mathrm{s^{-1}}$)', fontsize=35)
plt.ylabel ( '$\mathrm{X_{m}}$ ($\mathrm{Km}$ $\mathrm{s^{-1}}$)', fontsize=35)
#plt.title ('Homogeneous', fontsize =20)
plt.tick_params(axis='both', which='major', labelsize=18)

plt.xlim([-10, 310])
plt.ylim([-1.5, 230])

savefig('maximumVvsVmax.png')
```
