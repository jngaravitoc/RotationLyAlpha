```python
%pylab inline
import numpy as np 
import pylab as plt
import math
```

```python
def histo(data_dir, filename, path): 

    datoslyman = np.loadtxt(data_dir+filename, comments = '#')

    initpos=datoslyman[:,0]
    index_clean = np.where(~np.isnan(initpos))
    datoslyman = datoslyman[index_clean[0],:]

    x = datoslyman[:,6]
    scatt = datoslyman[:,7]
    hist, bins_edges = np.histogram(x, bins=50, normed=True)
    plt.plot(bins_edges[:-1] ,hist)
    out = open(path, 'w')
    n_points = np.size(hist)

    for i in range(n_points):
        out.write("%f %f \n" % (bins_edges[i],hist[i]))
    out.close()
    
    #return hist, bins_edges
```

```python
histo("../data/Homogeneous/", "Rotating200Sphere5tHOM.ascii", "../data/Homogeneous_Histograms/HistogramRotating200Sphere5tHOM")
histo("../data/Central/", "Rotating200Sphere5t.ascii", "../data/Central_Histograms/HistogramRotating200Sphere5t")
```
