```python
def J_surface(x, b, phi, i_angle, v_rot,v_th=12.6,tau0=1E7,R=1.0):
    a = 4.7E-4*(v_th/12.6)
    beta = arctan(b*sin(phi)/sqrt(R**2 - b**2))
    x_b = (v_rot/v_th) * sqrt(1.0- (b/R)**2 * (cos(phi)**2)) * cos(i_angle) * sin(beta)
    arg_in = fabs((x-x_b)**3)/(a*tau0)
    arg_in = arg_in * sqrt((2.0*pi**3)/27)
    final = (x-x_b)**2/(1.0+cosh(arg_in)) * (sqrt(pi)/(sqrt(24.0)*a*tau0))
    return final
```

```python
i_angle= 0.0
v_rot = 100.0
n_points = 100
x_array = linspace(-30.0,30.0, n_points)
phi_array = linspace(0.0,2.0*pi, 300)
b_array = linspace(0.0,0.999, 20)

delta_phi = phi_array[1]-phi_array[0]
delta_b = b_array[1] - b_array[0]

spectra = zeros(n_points)

for i in arange(n_points):
    integral = 0.0
    for b in b_array:
        for phi in phi_array:
            integrand = J_surface(x_array[i],b,phi,i_angle,v_rot) * b
            integral  = integral + (integrand * delta_phi * delta_b)
    
    spectra[i] = integral
    
```

```python
plot(x_array*12.6, spectra)
```
