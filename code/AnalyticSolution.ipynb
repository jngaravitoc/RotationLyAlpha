```python
%pylab inline
from subprocess import call
import os.path
```

Use the C source code to compute the analytical expectation

```python
tau = 1E7
v_rot_array = array([0.0, 100.0, 200.0, 300.0])
i_angle_array = array([0.0, 30.0, 45.0, 60.0, 90.0])

data_path = "../data/analytic"
for v_rot in v_rot_array:
    for i_angle in i_angle_array:
        fileout = "%s/analytic_logtau_%d_vrot_%d_angle_%d.dat"%(data_path, log10(tau), v_rot, i_angle)
        if( not os.path.exists(fileout)):
            print fileout
            command = "./analytic_solution.x %f %f %f > %s"%(tau, v_rot, i_angle, fileout)
            retcode = call(command,shell=True)
```

Load the Monte Carlo data

```python
MC_PX = 0
MC_PY = 1
MC_PZ = 2
MC_KX = 3
MC_KY = 4
MC_KZ = 5
MC_X = 6
MC_STATUS = 7
MC_NSCATT = 8
NCOLS=9

mc_data_path = "../data/Central"
n_vel = size(v_rot_array)
n_photons = 99360
full_mc_data = zeros((n_photons, NCOLS, n_vel))
for i  in arange(n_vel):
    fileout = "%s/Rotating%dSphere%dt.ascii"%(mc_data_path, v_rot_array[i], log10(tau))
    data = loadtxt(fileout)
    print fileout, shape(data)
    full_mc_data[0:n_photons,:,i] = data[0:n_photons,:] 
```

Compute the spectra from data for different angles and velocities

```python
n_vel = size(v_rot_array)
n_angles = size(i_angle_array)
n_x_items = 40
x_edges = linspace(-50.0,50.0, n_x_items+1)
x_center = (x_edges[:-1] + x_edges[1:])*0.5
mc_spectra = zeros((n_x_items, n_vel, n_angles))
```

```python
for i in arange(n_vel):
    for j in arange(n_angles):
        angle_rad = (90 - i_angle_array[j])*pi/180.0 #this is due to the different angle definition in the solution
        mu = cos(angle_rad)
        index = where((abs(full_mc_data[:,MC_KZ,i])<(mu+0.05))&(abs(full_mc_data[:,MC_KZ,i])>(mu-0.05)))
        index = index[0]
        tmp_x = full_mc_data[index,MC_X,i]
        histo,a = histogram(tmp_x, bins=x_edges, normed=1)
        mc_spectra[:,i,j] = histo[:]
```

```python
#plots varying the angle
plot(x_center, mc_spectra[:,3,0])
plot(x_center, mc_spectra[:,3,1])
plot(x_center, mc_spectra[:,3,2])
plot(x_center, mc_spectra[:,3,3])
```

```python
#plots varying the velocity
plot(x_center, mc_spectra[:,0,0])
plot(x_center, mc_spectra[:,1,0])
plot(x_center, mc_spectra[:,2,0])
plot(x_center, mc_spectra[:,3,0])
```

Load the analitic solutions

```python
n_x_analytic = 500
analytic_spectra = zeros((500, n_vel, n_angles))
x_center_analytic = zeros(500)
```

```python
for i in arange(n_vel):
    for j in arange(n_angles):
        filein = "%s/analytic_logtau_%d_vrot_%d_angle_%d.dat"%(data_path, log10(tau), v_rot_array[i], i_angle_array[j])
        data = loadtxt(filein)
        analytic_spectra[:,i,j] = data[:,1]
        x_center_analytic = data[:,0]
        delta_x = x_center_analytic[1] - x_center_analytic[0]
        norm_constant = sum(analytic_spectra[:,i,j]*delta_x)
        analytic_spectra[:,i,j] = data[:,1]/norm_constant
```

```python
fig = figure(1, figsize=(9.5,6.5))
ax = axes()
ax.set_xlabel("$\mathrm{v\ [km\ s^{-1}]}$",fontsize=25)
ax.set_ylabel("$\mathrm{flux\ [arbitrary\ units]}$",fontsize=25)
ticklabels_x = ax.get_xticklabels()
ticklabels_y = ax.get_yticklabels()

for label_x in ticklabels_x:
    label_x.set_fontsize(18)
    label_x.set_family('serif')
for label_y in ticklabels_y:
    label_y.set_fontsize(18)
    label_y.set_family('serif')

v_th = 12.4
plot(x_center_analytic*v_th, analytic_spectra[:,1,0]*100, color='black', linewidth=2)
plot(x_center_analytic*v_th, analytic_spectra[:,3,0]*100, color='black', linewidth=2)
scatter(x_center*v_th, mc_spectra[:,1,0]*100, marker='*', color='black',s=100, label='$V_{\mathrm{max}}=100\ \mathrm{km\ s^{-1}}$')
scatter(x_center*v_th, mc_spectra[:,3,0]*100, marker='o', color='black',s=100, label='$V_{\mathrm{max}}=300\ \mathrm{km\ s^{-1}}$')

ax.legend(loc=1, scatterpoints=1, prop={'size':18})
ax.set_title(r"$\tau_{H}=10^7\ \mu=0$", fontsize=25)

ax.set_xlim([-600.0,600])
ax.set_ylim([0.0,4.2])

filename='vary_vel'

savefig(filename + '.eps',format = 'eps', transparent=True)
savefig(filename + '.pdf',format = 'pdf', transparent=True)
savefig(filename + '.png',format = 'png', transparent=True)
```
