```python
#This notebook corresponds to section 4.  , Rotation effects...

from numpy import *
from pylab import *
import numpy as np
import math
import matplotlib.pyplot as plt

#1. Load the Data

data_path = "/home/Nicolas-Garavito/Dropbox/github/RotationLyAlpha/data/Homogeneous/"
filename = "Rotating100Sphere5tHOMDust.ascii"
filename3 = "Rotating300Sphere5tHOMDust.ascii"
data = np.loadtxt(data_path + filename, comments = '#')
data3 = np.loadtxt(data_path + filename3, comments = '#')

initpos = data[:,0]
index_clean = where(~np.isnan(initpos))
data = data[index_clean[0],:]

x = data[:,6]
#Nscatt = data[8]


initpos3 = data3[:,0]
index_clean3 = where(~np.isnan(initpos3))
data3 = data3[index_clean3[0],:]

x3 = data3[:,6]

print amin([x]), amax(x)
```
