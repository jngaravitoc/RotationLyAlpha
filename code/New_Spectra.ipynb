# Funcion que carga los datos

```python
def loading_data(filename, distribution):
    if distribution == 0:
        data = loadtxt("../data/Central/" + filename)
        initpos=data[:,0]
        index_clean = np.where(~np.isnan(initpos))
        data = data[index_clean[0],:]

    elif distribution == 1:
        data = loadtxt("../data/Homogeneous/" + filename)
        initpos=data[:,0]
        index_clean = np.where(~np.isnan(initpos))
        data = data[index_clean[0],:]

    kz = data[:,5]
    x = data[:,6]
    nscatt = data[:,8]
      
    return kz, x
```

# Funcion que hace los histogramas

```python
def histo(name, distribution, velocity):
    kz, x = loading_data(name, distribution)
    figure(figsize=(10.5, 9))
    histo = hist2d(x, abs(kz), bins=(120 , 120), normed=True)
    xlabel("$x$", fontsize=25)
    ylabel(r"$\mathrm{|k_{z}|}$", fontsize=25)
    plt.tick_params(axis='both', which='major', labelsize=18)
    X = amin(x) + 0.15*abs(amin(x))
    text(X, 0.9, velocity + "km/s", fontsize=25, color = 'White')
    #colorbar()
    savefig(name+".eps")
    
```
