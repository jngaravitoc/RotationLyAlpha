```python
from numpy import *
from pylab import *
import numpy as np
import math
import matplotlib.pyplot as plt

data_path = "/home/Nicolas-Garavito/"
data_dir="Dropbox/github/RotationLyAlpha/data/Central/"
data_dirm="Dropbox/github/RotationLyAlpha/data/MarkDijkstraModel/"

filenamej0="HistogramRotating0Sphere7t.ascii"
filenamej100 = 'HistogramRotating100Sphere7t.ascii'
filenamej300 = 'HistogramRotating300Sphere7t.ascii'

filenamem0="spec.vrot0.bin"
filenamem100 = 'spec.vrot100.bin'
filenamem300 = 'spec.vrot300.bin'


datoslymanj0 = np.loadtxt(data_path+data_dir+filenamej0, comments = '#')
datoslymanj100 = np.loadtxt(data_path+data_dir+filenamej100, comments = '#')
datoslymanj300 = np.loadtxt(data_path+data_dir+filenamej300, comments = '#')

datoslymanm0 = np.loadtxt(data_path+data_dirm+filenamem0, comments = '#')
datoslymanm100 = np.loadtxt(data_path+data_dirm+filenamem100, comments = '#')
datoslymanm300 = np.loadtxt(data_path+data_dirm+filenamem300, comments = '#')

xj0 = datoslymanj0[:,0]
yj0 = datoslymanj0[:,1]

xj100 = datoslymanj100[:,0]
yj100 = datoslymanj100[:,1]

xj300 = datoslymanj300[:,0]
yj300 = datoslymanj300[:,1]


xm0 = datoslymanm0[:,0]
ym0 = datoslymanm0[:,1]

xm100 = datoslymanm100[:,0]
ym100 = datoslymanm100[:,1]

xm300 = datoslymanm300[:,0]
ym300 = datoslymanm300[:,1]


```

```python
pj0, = plot(xj0,yj0, '-b')
pj100, = plot(xj100,yj100, '-b')
pj300, = plot(xj300,yj300, '-b')

pm0, = plot(xm0,ym0, '--k')
pm100, = plot(xm100,ym100, '--k')
pm300, = plot(xm300,ym300, '--k')

legend([pj0, pj100, pj300, pm0, pm100, pm300],[r'$V_{CLARA}=0kms^{-1}$',r'$V_{CLARA}=100kms^{-1}$',r'$V_{CLARA}=300kms^{-1}$',r'$V_{MC}=0kms^{-1}$',r'$V_{MC}=100kms^{-1}$',r'$V_{MC}=300kms^{-1}$'])

ylabel(r'$J(x)$', fontsize=20)
xlabel(r'$x$', fontsize=20)

xlim([-60, 100])

#savefig('CLARAandMark2.png')

plt.show()
```
