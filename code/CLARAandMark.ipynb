```python
from numpy import *
from pylab import *
import numpy as np
import math
import matplotlib.pyplot as plt

data_path = "/home/nicolas/"
data_dir="Dropbox/github/RotationLyAlpha/data/Central/"
data_dirh="Dropbox/github/RotationLyAlpha/data/Homogeneous/"
data_dirm="Dropbox/github/RotationLyAlpha/data/MarkDijkstraModel/"

filenamej0="HistogramRotating0Sphere7t.ascii"
filenamej50 = "HistogramRotating50Sphere7t.ascii"
filenamej100 = 'HistogramRotating100Sphere7t.ascii'
filenamej200 = 'HistogramRotating200Sphere7t.ascii'
filenamej300 = 'HistogramRotating300Sphere7t.ascii'
filenamej300 = 'HistogramRotating300Sphere7t.ascii'
filenamej300HOM = 'HistogramRotating300Sphere7tHOM.ascii'

filenamem0="spec.vrot0.bin"
filenamem50 = "spec.vrot50.bin"
filenamem100 = 'spec.vrot100.bin'
filenamem200 = 'spec.vrot200.bin'
filenamem300 = 'spec.vrot300.bin'
filenamem300sqrt = 'spec.vrot300.sqrtpi.bin'
filenamem300sqrt2 = 'spec.vrot300.sqrtpiv2.out'
filenamem300HOM = 'spec.uniform.vrot300.sqrtpi.bin'


datoslymanj0 = np.loadtxt(data_path+data_dir+filenamej0, comments = '#')
datoslymanj50 = np.loadtxt(data_path+data_dir+filenamej50, comments = '#')
datoslymanj100 = np.loadtxt(data_path+data_dir+filenamej100, comments = '#')
datoslymanj200 = np.loadtxt(data_path+data_dir+filenamej200, comments = '#')
datoslymanj300 = np.loadtxt(data_path+data_dir+filenamej300, comments = '#')
datoslymanj300HOM = np.loadtxt(data_path+data_dirh+filenamej300HOM, comments = '#')

datoslymanm0 = np.loadtxt(data_path+data_dirm+filenamem0, comments = '#')
datoslymanm50 = np.loadtxt(data_path+data_dirm+filenamem50, comments = '#')
datoslymanm100 = np.loadtxt(data_path+data_dirm+filenamem100, comments = '#')
datoslymanm200 = np.loadtxt(data_path+data_dirm+filenamem200, comments = '#')
datoslymanm300 = np.loadtxt(data_path+data_dirm+filenamem300, comments = '#')
datoslymanm300sqrt = np.loadtxt(data_path+data_dirm+filenamem300sqrt, comments = '#')
datoslymanm300sqrt2 = np.loadtxt(data_path+data_dirm+filenamem300sqrt2, comments = '#')
datoslymanm300HOM = np.loadtxt(data_path+data_dirm+filenamem300HOM, comments = '#')

xj0 = datoslymanj0[:,0]
yj0 = datoslymanj0[:,1]

xj50 = datoslymanj50[:,0]
yj50 = datoslymanj50[:,1]

xj100 = datoslymanj100[:,0]
yj100 = datoslymanj100[:,1]

xj200 = datoslymanj200[:,0]
yj200 = datoslymanj200[:,1]

xj300 = datoslymanj300[:,0]
yj300 = datoslymanj300[:,1]

xj300HOM = datoslymanj300HOM[:,0]
yj300HOM = datoslymanj300HOM[:,1]

xm0 = datoslymanm0[:,0]
ym0 = datoslymanm0[:,1]
deltam0 = sum(ym0*abs(xm0[0]-xm0[1]))

xm50 = datoslymanm50[:,0]
ym50 = datoslymanm50[:,1]
deltam50 = sum(ym50*abs(xm50[0]-xm50[1]))

xm100 = datoslymanm100[:,0]
ym100 = datoslymanm100[:,1]
deltam100 = sum(ym100*abs(xm100[0]-xm100[1]))

xm200 = datoslymanm200[:,0]
ym200 = datoslymanm200[:,1]
deltam200 = sum(ym200*abs(xm200[0]-xm200[1]))

xm300 = datoslymanm300[:,0]
ym300 = datoslymanm300[:,1]
deltam300 = sum(ym300*abs(xm300[0]-xm300[1]))

xm300sqrt = datoslymanm300sqrt[:,0]
ym300sqrt = datoslymanm300sqrt[:,1]
deltam300sqrt = sum(ym300sqrt*abs(xm300sqrt[0]-xm300sqrt[1]))

xm300sqrt2 = datoslymanm300sqrt2[:,0]
ym300sqrt2 = datoslymanm300sqrt2[:,1]
deltam300sqrt2 = sum(ym300sqrt2*abs(xm300sqrt2[0]-xm300sqrt2[1]))

xm300sqrtHOM = datoslymanm300HOM[:,0]
ym300sqrtHOM = datoslymanm300HOM[:,1]
deltam300sqrtHOM = sum(ym300sqrtHOM*abs(xm300sqrtHOM[0]-xm300sqrtHOM[1]))
```

```python
pj0, = plot(xj0,yj0, '-b')

pm0, = plot(xm0,ym0/deltam0, '-k')


legend([pj0, pm0],['CLARA','Mark'])

title(r'$V=0kms^{-1}, \tau_{H}=10^{7}, T=10^4K$, Central', fontsize =20)
ylabel(r'$J(x)$', fontsize=20)
xlabel(r'$x$', fontsize=20)

xlim([-60, 60])

savefig('CLARAandMark0.png')

plt.show()
```
