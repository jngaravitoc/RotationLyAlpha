```python
def ef(filename, mode):
    if mode == 1:
        data_dir="../data/Homogeneous/" # Select the homogeneous or central folder  
    elif mode == 0:
        data_dir ="../data/Central/"
    datoslyman = np.loadtxt(data_dir+filename, comments = '#')

    initpos=datoslyman[:,0]
    index_clean = where(~np.isnan(initpos))
    datoslyman = datoslyman[index_clean[0],:]

    scatt = datoslyman[:,7]
     
    x1 = float(len(scatt))
    
    index = np.where(scatt==0)

    scatt = scatt[index]
    
    x2 = float(len(scatt))
        
    ef = x2/x1
    return ef

```
