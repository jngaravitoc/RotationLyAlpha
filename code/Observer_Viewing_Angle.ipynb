```python
def get_mu_histo(filename, theta=pi/2.0):
    data = loadtxt(filename)
    kx = data[:, 3]
    ky = data[:, 4]
    kz = data[:, 5]
    R = np.array([0, sin(theta), cos(theta)])
    miu = kx*R[0] + ky*R[1] + kz*R[2]
    n_bins = 12
    delta_mu = 2.0/n_bins
    histo, bin_edges = histogram(miu, bins=n_bins)
    histo = 2.0*histo/(delta_mu*size(miu))
    return [histo, bin_edges]
```

```python
filename='../data/Homogeneous/Rotating0Sphere5tHOM.ascii'
histo0, bin_edges0 = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating100Sphere5tHOM.ascii'
histo1, bin_edges1 = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating200Sphere5tHOM.ascii'
histo2, bin_edges2 = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating300Sphere5tHOM.ascii'
histo3, bin_edges3 = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating0Sphere6tHOM.ascii'
histo4, bin_edges4 = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating100Sphere6tHOM.ascii'
histo5, bin_edges5 = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating200Sphere6tHOM.ascii'
histo6, bin_edges6 = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating300Sphere6tHOM.ascii'
histo7, bin_edges7 = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating0Sphere7tHOM.ascii'
histo8, bin_edges8 = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating100Sphere7tHOM.ascii'
histo9, bin_edges9 = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating200Sphere7tHOM.ascii'
histo10, bin_edges10 = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating300Sphere7tHOM.ascii'
histo11, bin_edges11 = get_mu_histo(filename)
```
