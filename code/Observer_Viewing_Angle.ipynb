```python
def get_mu_histo(filename, theta=pi/2.0):
    data = loadtxt(filename)
    kx = data[:, 3]
    ky = data[:, 4]
    kz = data[:, 5]
    R = np.array([0, sin(theta), cos(theta)])
    miu = kx*R[0] + ky*R[1] + kz*R[2]
    n_bins = 12
    delta_mu = 2.0/n_bins
    histo, bin_edges = histogram(miu, bins=n_bins)
    histo = 2.0*histo/(delta_mu*size(miu))
    return [histo, bin_edges]
```

```python
filename='../data/Homogeneous/Rotating0Sphere5tHOM.ascii'
histo0, bin_edges = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating100Sphere5tHOM.ascii'
histo1, bin_edges = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating200Sphere5tHOM.ascii'
histo2, bin_edges = get_mu_histo(filename)

filename='../data/Homogeneous/Rotating300Sphere5tHOM.ascii'
histo3, bin_edges = get_mu_histo(filename)

figure(num=None, figsize=(9.5,9.5))
plt.title('$\mathrm{Homogeneous}$', fontsize=27)
plt.xlabel(r'$\mathrm{\mu}$', fontsize=27)
plt.ylabel(r'$\mathrm{F(\mu)}$', fontsize=27)
plt.tick_params(axis='both', which='major', labelsize=18)
plt.ylim([0.980, 1.028])

plot(bin_edges[0:-1], histo0, 'k', linewidth=0.8)
plot(bin_edges[0:-1], histo1, 'k', linewidth=1.5)
plot(bin_edges[0:-1], histo2, 'k', linewidth=2.2)
plot(bin_edges[0:-1], histo3, 'k', linewidth=2.9)
legend([r'$\mathrm{0}$ $\mathrm{kms^{-1}}$', r'$\mathrm{100}$ $\mathrm{kms^{-1}}$', r'$\mathrm{200}$ $\mathrm{kms^{-1}}$', r'$\mathrm{300}$ $\mathrm{kms^{-1}}$'], fontsize=20)
savefig('mu_homogeneous.png')
```
