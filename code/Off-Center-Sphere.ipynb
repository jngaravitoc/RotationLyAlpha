```python
%pylab inline
from numpy import *
from pylab import *
import numpy as np
import math
import matplotlib.pyplot as plt

data_dir="../data/Homogeneous/"
#filename200="V3007tHOM_in.ascii"
#filenameOUT200 = 'V3007tHOM_out.ascii'

filename200='V3006tHOM_in.ascii'
filenameOUT200 = 'V3006tHOM_out.ascii'


#filename50="V507tHOM_in.ascii"
#filenameOUT50 = 'V507tHOM_out.ascii'

filename50="V50HOM6t_in.ascii"
filenameOUT50 = 'V50HOM6t_out.ascii'

datoslyman200 = np.loadtxt(data_dir+filename200, comments = '#')
datoslymanOUT200 = np.loadtxt(data_dir+filenameOUT200, comments = '#')

datoslyman50 = np.loadtxt(data_dir+filename50, comments = '#')
datoslymanOUT50 = np.loadtxt(data_dir+filenameOUT50, comments = '#')


initpos=datoslymanOUT50[:,0]
index_clean = where(~np.isnan(initpos))
datoslymanOUT50 = datoslymanOUT50[index_clean[0],:]
datoslyman50 = datoslyman50[index_clean[0],:]


initpos=datoslymanOUT200[:,0]
index_clean = where(~np.isnan(initpos))
datoslymanOUT200 = datoslymanOUT200[index_clean[0],:]
datoslymanOUT200 = datoslymanOUT200[index_clean[0],:]

x50 = datoslymanOUT50[:,6]

X50 = datoslyman50[:,0]
Y50 = datoslyman50[:,1]
Z50 = datoslyman50[:,2]


kx50 = datoslymanOUT50[:,3]
ky50 = datoslymanOUT50[:,4]
kz50 = datoslymanOUT50[:,5]


x200 = datoslymanOUT200[:,6]

X200 = datoslyman200[:,0]
Y200 = datoslyman200[:,1]
Z200 = datoslyman200[:,2]


kx200 = datoslymanOUT200[:,3]
ky200 = datoslymanOUT200[:,4]
kz200 = datoslymanOUT200[:,5]


obs = array([1.0,0.0,0.0])
weight50 = kx50*obs[0] +  ky50*obs[1] + kz50*obs[2]
weight50[weight50<0] = 0.0

norm50 = sqrt(kx50**2 + ky50**2 + kz50**2)
norm_r50 = amax(sqrt(X50**2 + Y50**2 + Z50**2))
print "norma", norm_r50
X50 = X50/norm_r50
Y50 = Y50/norm_r50
Z50 = Z50/norm_r50

weight200 = kx200*obs[0] +  ky200*obs[1] + kz200*obs[2]
weight200[weight200<0] = 0.0

norm200 = sqrt(kx200**2 + ky200**2 + kz200**2)
norm_r200 = amax(sqrt(X200**2 + Y200**2 + Z200**2))
X200 = X200/norm_r200
Y200 = Y200/norm_r200
Z200 = Z200/norm_r200


print amax(X50), amax(Y50), amin(X50), amin(Y50), amin(weight50)
```

```python
def offcenter(xmin, xmax, ymin, ymax, zmin, zmax, XV, YV, ZV, xV, V, weightV):
    selected = where((XV >= xmin) &  (XV <= xmax) & 
                    (ZV >= zmin) & (ZV <=zmax) & 
                    (YV >= ymin) & (YV <= ymax))
    xV_selected = xV[selected]
    new_XV = XV[selected]
    new_YV = YV[selected]
    new_ZV = ZV[selected]
    new_wV = weightV[selected]
    return [xV_selected, new_XV, new_YV, new_ZV, new_wV]
                      
x50_new_1, new_x50_1, new_y50_1, new_z50_1, new_w50_1 = offcenter(0.0, 0.2, 0.5, 1.0, -0.5, 0.5, X50, Y50, Z50, x50, 50, weight50)
x50_new_2, new_x50_2, new_y50_2, new_z50_2, new_w50_2 = offcenter(0.0, 0.2, -1.0, -0.5, -0.5, 0.5, X50, Y50, Z50, x50, 50, weight50)
                      
x200_new_1, new_x200_1, new_y200_1, new_z200_1, new_w200_1 = offcenter(0.0, 0.2, 0.5, 1.0, -0.5, 0.5, X200, Y200, Z200, x200, 200, weight200)   
x200_new_2, new_x200_2, new_y200_2, new_z200_2, new_w200_2 = offcenter(0.0, 0.2, -1.0, -0.5, -0.5, 0.5, X200, Y200, Z200, x200, 200, weight200)

print size(x50_new_1)
```
