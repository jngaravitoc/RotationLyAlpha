### Uploading data with $N_{\gamma} = 10^5$

```python
def definig_data(v, tau):
    #data path
    data_dir = "../data/Homogeneous/"
    filename_v_tau = "V"+v+tau+"HOM_in.ascii"
    filenameOUT_v_tau = "V"+v+tau+"HOM_out.ascii"
    
    #cleaning data
    datoslyman_v_tau = np.loadtxt(data_dir+filename_v_tau, comments = '#')
    datoslymanOUT_v_tau = np.loadtxt(data_dir+filenameOUT_v_tau, comments = '#')
    initpos=datoslymanOUT_v_tau[:,0]
    index_clean = where(~np.isnan(initpos))
    datoslymanOUT_v_tau = datoslymanOUT_v_tau[index_clean[0],:]
    datoslyman_v_tau = datoslyman_v_tau[index_clean[0],:]
    
    #Defining cuantites
    x_v_tau = datoslymanOUT_v_tau[:,6]
    X_v_tau = datoslyman_v_tau[:,0]
    Y_v_tau = datoslyman_v_tau[:,1]
    Z_v_tau = datoslyman_v_tau[:,2]
    kx_v_tau = datoslymanOUT_v_tau[:,3]
    ky_v_tau = datoslymanOUT_v_tau[:,4]
    kz_v_tau = datoslymanOUT_v_tau[:,5]
    
    norm = amax(np.sqrt(X_v_tau**2 + Y_v_tau**2 + Z_v_tau**2))
    X_v_tau = X_v_tau / norm
    Y_v_tau = Y_v_tau / norm
    Z_v_tau = Z_v_tau / norm
    
    return (x_v_tau, X_v_tau, Y_v_tau, Z_v_tau, kx_v_tau, ky_v_tau, kz_v_tau)

x_50_5t, X_50_5t, Y_50_5t, Z_50_5t, kx_50_5t, ky_50_5t, kz_50_5t = definig_data('50', '5t')

x_100_5t, X_100_5t, Y_100_5t, Z_100_5t, kx_100_5t, ky_100_5t, kz_100_5t = definig_data('100', '5t')
x_200_5t, X_200_5t, Y_200_5t, Z_200_5t, kx_200_5t, ky_200_5t, kz_200_5t = definig_data('200', '5t')
x_300_5t,X_300_5t,Y_300_5t, Z_300_5t, kx_300_5t, ky_300_5t, kz_300_5t = definig_data('300', '5t') 

x_50_6t, X_50_6t, Y_50_6t, Z_50_6t, kx_50_6t, ky_50_6t, kz_50_6t = definig_data('50', '6t')
x_100_6t, X_100_6t, Y_100_6t, Z_100_6t, kx_100_6t, ky_100_6t, kz_100_6t = definig_data('100', '6t')
x_200_6t, X_200_6t, Y_200_6t, Z_200_6t, kx_200_6t, ky_200_6t, kz_200_6t = definig_data('200', '6t')
x_300_6t, X_300_6t, Y_300_6t, Z_300_6t, kx_300_6t, ky_300_6t, kz_300_6t = definig_data('300', '6t')

x_50_7t, X_50_7t, Y_50_7t, Z_50_7t, kx_50_7t, ky_50_7t, kz_50_7t = definig_data('50', '7t')
x_100_7t, X_100_7t, Y_100_7t, Z_100_7t, kx_100_7t, ky_100_7t, kz_100_7t = definig_data('100', '7t')
x_200_7t, X_200_7t, Y_200_7t, Z_200_7t, kx_200_7t, ky_200_7t, kz_200_7t = definig_data('200', '7t')
x_300_7t, X_300_7t, Y_300_7t, Z_300_7t, kx_300_7t, ky_300_7t, kz_300_7t = definig_data('300', '7t')



```

### The following function select the photons locate in the off-centred distribution showed in Fig.8

```python
def offcenter_ZZ(xV, XV, YV, ZV, kxV, kyV, kzV, V, R, D2R):
    selected = where( ((ZV <= np.sqrt(R**2 - (XV-D2R)**2 - YV**2))  & (ZV >= 0.0) ) |
                    ( (ZV >= -np.sqrt(R**2 - (XV-D2R)**2 - YV**2))  & (ZV <= 0.0)) )
    
    xV_selected = xV[selected]
    new_XV = XV[selected]
    new_YV = YV[selected]
    new_ZV = ZV[selected]
    #new_wV = weightV[selected]
    new_xV = xV[selected]
    new_kxV = kxV[selected]
    new_kyV = kyV[selected]
    new_kzV = kzV[selected]
    
    return [xV_selected, new_XV, new_YV, new_ZV, new_kxV, new_kyV, new_kzV, new_xV]
```

### Now we make a Histogram of $\mu = \hat{k_{out}} \cdot \hat{k_{obs}}$

### where $\hat{k_{obs}}$  represents the direction of the line of sight of the observer

```python
# ZZ implementation
def get_mu_histo(kx, ky, kz, theta):
    R = np.array([0, sin(theta), cos(theta)])
    mu = kx*R[0] + ky*R[1] + kz*R[2]
    n_bins = 12
    delta_mu = 2.0/n_bins
    histo, bin_edges = histogram(mu, bins=n_bins)
    histo = 2.0*histo/(delta_mu*size(mu))
    return [histo, bin_edges, mu]
```
