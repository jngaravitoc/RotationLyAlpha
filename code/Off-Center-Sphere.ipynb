```python
#Fix this function, in order to write the data.
def offcenter(xmin, xmax, ymin, ymax, zmin, zmax, XV, YV, ZV, xV, V, weightV, R):
    selected = where((XV <= np.sqrt((R**2) - (ymin**2) - (zmax**2)) + 0.1) & (XV >= np.sqrt((R**2) - (ymax**2) - (zmax**2)) + 0.1)
                    &(YV <= np.sqrt((R**2) - ((xmin-0.1)**2) - (zmax**2)) + 0.1) & (YV >= np.sqrt((R**2) - ((xmin + 0.1)**2) - (zmax**2)) + 0.1))
    xV_selected = xV[selected]
    new_XV = XV[selected]
    new_YV = YV[selected]
    new_ZV = ZV[selected]
    new_wV = weightV[selected]
    new_x50 = x50[selected]
    return [xV_selected, new_XV, new_YV, new_ZV, new_wV, new_x50]
                      
x50_new_1, new_x50_1, new_y50_1, new_z50_1, new_w50_1, new1_x50 = offcenter(-0.1, 0.1, 0.0, 0.1, -0.1, 0.1, X50, Y50, Z50, x50, 50, weight50, 0.2)
#x50_new_2, new_x50_2, new_y50_2, new_z50_2, new_w50_2, new2_x50 = offcenter(-0.2, 0.2, 0.3, 0.6, -0.2, 0.2, X50, Y50, Z50, x50, 50, weight50, 0.3)
#x50_new_3, new_x50_3, new_y50_3, new_z50_3, new_w50_3, new3_x50 = offcenter(-0.2, 0.2, 0.6, 0.9, -0.2, 0.2, X50, Y50, Z50, x50, 50, weight50, 0.3)


```
