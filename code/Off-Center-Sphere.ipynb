```python
%pylab inline
from numpy import *
from pylab import *
import numpy as np
import math
import matplotlib.pyplot as plt

data_dir="../data/Homogeneous/"
filename="V2007tHOM_in.ascii"
filenameOUT = 'V2007tHOM_out.ascii'

datoslyman = np.loadtxt(data_dir+filename, comments = '#')
datoslymanOUT = np.loadtxt(data_dir+filenameOUT, comments = '#')

initpos=datoslymanOUT[:,0]
index_clean = where(~np.isnan(initpos))
datoslymanOUT = datoslymanOUT[index_clean[0],:]

x = datoslymanOUT[:,6]

X = datoslyman[:,0]
Y = datoslyman[:,1]
Z = datoslyman[:,2]


kx = datoslymanOUT[:,3]
ky = datoslymanOUT[:,4]
kz = datoslymanOUT[:,5]

obs = array([1.0,0.0,0.0])
weight = kx*obs[0] +  ky*obs[1] + kz*obs[2]
weight[weight<0] = 0.0

norm = sqrt(kx**2 + ky**2 + kz**2)
norm_r = sqrt(X**2 + Y**2 + Z**2)
X = X/norm_r
Y = Y/norm_r
Z = Z/norm_r

print amax(X), amax(Y), amin(X), amin(Y), amin(weight)
```

```python
scatter(X,Y)
```

```python
#Fix this function, in order to write the data.
def offcenter(xmin, xmax, ymin, ymax, zmin, zmax):
    selected = where((X >= xmin) &  (X <= xmax) & 
                    (Z >= zmin) & (Z <=zmax) & 
                    (Y >= ymin) & (Y <= ymax))
    x_selected = x[selected]
    new_X = X[selected]
    new_Y = Y[selected]
    new_Z = Z[selected]
    new_w = weight[selected]
    return [x_selected, new_X, new_Y, new_Z, new_w]
                      
x_new_1, new_x_1, new_y_1, new_z_1, new_w_1 = offcenter(0.0, 1.0, 0.0, 1.0, -0.5,0.5)
x_new_2, new_x_2, new_y_2, new_z_2, new_w_2 = offcenter(0.0, 1.0, -1.0, -0.0, -0.5,0.5)
```

```python
scatter(new_x_1, new_w_1)
```

```python
a_1 = x_new_1
freqbin = arange(amin(a_1), amax(a_1), 4.0)
hist_1, bin_edges_1 = np.histogram(a_1,bins = freqbin, density=True, weights=new_w_1)

a_2 = x_new_2
freqbin = arange(amin(a_2), amax(a_2), 4.0)
hist_2, bin_edges_2 = np.histogram(a_2,bins = freqbin, density=True, weights=new_w_2)




p0, = plot(bin_edges_1[1:],hist_1)
p0, = plot(bin_edges_2[1:],hist_2)

plt.xlabel (r'$x$', fontsize=20)
plt.ylabel ('Intensity (Arbitrary Units)', fontsize=20)
plt.title(r"Inhomogeneous, $V_{max}=200Kms^{-1}$,$\tau_{H}=10^{7}, datos3 $",  fontsize=20)
savefig('inhomogenoeusdatos3.png')
```
