```python
# ZZ implementation
def get_mu_histo(kx, ky, kz, theta):
    R = np.array([0, sin(theta), cos(theta)])
    mu = kx*R[0] + ky*R[1] + kz*R[2]
    n_bins = 12
    delta_mu = 2.0/n_bins
    histo, bin_edges = histogram(mu, bins=n_bins)
    histo = 2.0*histo/(delta_mu*size(mu))
    return [histo, bin_edges, mu]

a, b, c, = get_mu_histo(new_kx50_1, new_ky50_1, new_kz50_1, 0.0)
a1, b1, c1, = get_mu_histo(new_kx50_4, new_ky50_4, new_kz50_4, 0.0)

figure(num=None, figsize=(9.5,9.5))

plot(b[0:-1], a, color = 'k', linewidth=1.0)
plot(b1[0:-1], a1, color = 'k', linewidth=3.0)

plt.title('$\mathrm{Homogeneous}$', fontsize=35)
plt.xlabel(r'$\mathrm{\mu}$', fontsize=35)
plt.ylabel(r'$\mathrm{F(\mu)}$', fontsize=35)
plt.tick_params(axis='both', which='major', labelsize=18)
plt.ylim([0.92, 1.098])
plt.legend(['$\mathrm{R=0.25}$', '$\mathrm{R=0.5}$'], fontsize=20)

savefig('mu_OC.png')
```
