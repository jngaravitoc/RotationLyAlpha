```python
# ZZ implementation
def get_mu_histo(kx, ky, kz, theta):
    R = np.array([0, sin(theta), cos(theta)])
    mu = kx*R[0] + ky*R[1] + kz*R[2]
    n_bins = 12
    delta_mu = 2.0/n_bins
    histo, bin_edges = histogram(mu, bins=n_bins)
    histo = 2.0*histo/(delta_mu*size(mu))
    return [histo, bin_edges, mu]

get_mu_histo(new_kx50_1, new_ky50_1, new_kz50_1, pi/2.0)
get_mu_histo(new_kx50_2, new_ky50_2, new_kz50_2, pi/2.0)
get_mu_histo(new_kx50_3, new_ky50_3, new_kz50_3, pi/2.0)
get_mu_histo(new_kx50_4, new_ky50_4, new_kz50_4, pi/2.0)
```

```python
def get_mu(X, Y, Z, xV):
    x, y, z = get_mu_histo(X, Y, Z, pi/2.0)
    selected = where((z>-0.1) & (z<0.1))
    new_xV = xV[selected]
    return new_xV

new_x50_1 = get_mu(new_kx50_1, new_ky50_1, new_kz50_1, x50_new_1)
new_x50_2 = get_mu(new_kx50_2, new_ky50_2, new_kz50_2, x50_new_2)
new_x50_3 = get_mu(new_kx50_3, new_ky50_3, new_kz50_3, x50_new_3)
new_x50_4 = get_mu(new_kx50_4, new_ky50_4, new_kz50_4, x50_new_4)
```
