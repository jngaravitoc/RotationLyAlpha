```python
#Fix this function, in order to write the data.
def offcenter(xmin, xmax, ymin, ymax, zmin, zmax, XV, YV, ZV, xV, V, weightV):
    selected = where((XV >= xmin) &  (XV <= xmax) & 
                    (ZV >= zmin) & (ZV <=zmax) & 
                    (YV >= ymin) & (YV <= ymax))
    xV_selected = xV[selected]
    new_XV = XV[selected]
    new_YV = YV[selected]
    new_ZV = ZV[selected]
    new_wV = weightV[selected]
    return [xV_selected, new_XV, new_YV, new_ZV, new_wV]
                      
x50_new_1, new_x50_1, new_y50_1, new_z50_1, new_w50_1 = offcenter(-0.2, 0.2, 0.0, 0.2, -0.2, 0.2, X50, Y50, Z50, x50, 50, weight50)
x50_new_2, new_x50_2, new_y50_2, new_z50_2, new_w50_2 = offcenter(-0.2, 0.2, 0.2, 0.4, -0.2, 0.2, X50, Y50, Z50, x50, 50, weight50)
x50_new_3, new_x50_3, new_y50_3, new_z50_3, new_w50_3 = offcenter(-0.2, 0.2, 0.4, 0.6, -0.2, 0.2, X50, Y50, Z50, x50, 50, weight50)
x50_new_4, new_x50_4, new_y50_4, new_z50_4, new_w50_4 = offcenter(-0.2, 0.2, 0.6, 0.8, -0.2, 0.2, X50, Y50, Z50, x50, 50, weight50)
x50_new_5, new_x50_5, new_y50_5, new_z50_5, new_w50_5 = offcenter(-0.2, 0.2, 0.8, 1.0, -0.2, 0.2, X50, Y50, Z50, x50, 50, weight50)


print size(x50_new_1)
```
