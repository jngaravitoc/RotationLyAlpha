```python
data_dirh="../data/Homogeneous/"
data_dirc="../data/Central/"
filenameh0="Rotating0Sphere5tHOM.ascii"
filenameh300="Rotating300Sphere5tHOM.ascii"
filenamec0="Rotating0Sphere5t.ascii"
filenamec300="Rotating300Sphere5t.ascii"

datoslymanh0 = np.loadtxt(data_dirh+filenameh0, comments = '#')
datoslymanh300 = np.loadtxt(data_dirh+filenameh300, comments = '#')
datoslymanc0 = np.loadtxt(data_dirc+filenamec0, comments = '#')
datoslymanc300 = np.loadtxt(data_dirc+filenamec300, comments = '#')
```

```python
x = datoslymanh0[:,6]
scatt = datoslymanh0[:,8]

#plt.autoscale_on=True

#plt.ylim([0,16])

figure(num=None, figsize=(9,9.5))
#axis = plt.axes() 
xmin = -15.0
xmax = 15.0
ymin = 0.0
ymax = 6.0

#ax1 = subplot(1, 1, 1)
hist,xedges,yedges = numpy.histogram2d(log10(scatt),x,bins=(30,40),range=[[ymin,ymax],[xmin,xmax]])
aspectratio = 1.0*(xmax - xmin)/(1.0*ymax - ymin)
imshow(log10(hist+1),extent=[xmin,xmax,ymin,ymax],interpolation='nearest',origin='lower', aspect=aspectratio)

plt.ylim([ymin,ymax])
plt.xlabel("$x$",fontsize=35)
plt.ylabel("$\log_{10} \mathrm{ N_{scatt}}$",fontsize=35)
#ax1.tick_params(axis='both', which='major', labelsize=18)

#new_tick_locations = np.array([0, .166, .3333, .5, .666, .833, 1])


#ax2 = ax1.twiny()
#ax2.set_xticks(new_tick_locations)
#ax2.tick_params(labelsize=18)
#ax2.set_xticklabels(['-768','-512','-256', '0', '256', '512', '768'])
plt.text(-19, 6.2, r"$\mathrm{Homogeneous}$, $\mathrm{\tau_{H}=10^5}, \mathrm{V_{max}=0 km}$ $\mathrm{s^{-1}}$", fontsize=35)

# Color bar
cbar = colorbar(shrink = 0.75, orientation = 'vertical')
cbar.ax.tick_params(labelsize = 15) 
cbar.ax.xaxis.get_offset_text().set_position((1.5,1))
#cbar()
savefig ('2dHistogram0t5HOM.png')
show()
```
