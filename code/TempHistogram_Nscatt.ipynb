```python
%pylab inline
from pylab import *
import numpy
```

```python
#data_dir="../data/raw_data/new_data/RUN_02/"
data_dir="../data/Central/"
#filename="V0hom5t.ascii"
filename="Rotating300Sphere5t.ascii"

datoslyman = np.loadtxt(data_dir+filename, comments = '#')

x = datoslyman[:,6]
scatt = datoslyman[:,8]

#plt.autoscale_on=True

#plt.ylim([0,16])

figure(num=None, figsize=(9.5,9))
axis = plt.axes() 
xmin = -35.0
xmax = 35.0
ymin = 0.0
ymax = 6.0

hist,xedges,yedges = numpy.histogram2d(log10(scatt),x,bins=(30,40),range=[[ymin,ymax],[xmin,xmax]])
aspectratio = 1.0*(xmax - xmin)/(1.0*ymax - ymin)
imshow(log10(hist+1),extent=[xmin,xmax,ymin,ymax],interpolation='nearest',origin='lower', aspect=aspectratio)
axis.set_ylim([ymin,ymax])
axis.set_title(r"$\mathrm{Homogeneous}$, $\tau=10^5, V=0 kms^{-1}$", fontsize=25)
axis.set_xlabel("$x$",fontsize=25)
axis.set_ylabel("$\log_{10} \mathrm{ N_{scatt}}$",fontsize=25)
colorbar()
savefig ('2dHistogram300t5.png')
show()
```
