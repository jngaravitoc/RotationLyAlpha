```python
from pylab import *
import numpy


data_path = "/home/Nicolas-Garavito/"
data_dir="Dropbox/github/RotationLyAlpha/data/raw_data/new_data/RUN_02/"
filename="V3005tHOM_out.ascii"

datoslyman = np.loadtxt(data_path+data_dir+filename, comments = '#')

x = datoslyman[:,6]
scatt = datoslyman[:,8]

#plt.autoscale_on=True

plt.ylim([0,16])

hist,xedges,yedges = numpy.histogram2d(x,log(scatt),bins=30,range=[[-25,25],[0,16]])
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1] ]
imshow(hist.T,extent=extent,interpolation='nearest',origin='lower')
colorbar()
#savefig ('2dHistogram50t5HOM.png')
show()
```
