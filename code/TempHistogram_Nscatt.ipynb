```python
#data_dir="../data/raw_data/new_data/RUN_02/"
data_dir="../data/Homogeneous/"
#filename="V0hom5t.ascii"
filename="Rotating0Sphere5tHOM.ascii"

datoslyman = np.loadtxt(data_dir+filename, comments = '#')

x = datoslyman[:,6]
scatt = datoslyman[:,8]

#plt.autoscale_on=True

#plt.ylim([0,16])

figure(num=None, figsize=(9,9))
#axis = plt.axes() 
xmin = -15.0
xmax = 15.0
ymin = 0.0
ymax = 6.0

ax1 = subplot(1, 1, 1)
hist,xedges,yedges = numpy.histogram2d(log10(scatt),x,bins=(30,40),range=[[ymin,ymax],[xmin,xmax]])
aspectratio = 1.0*(xmax - xmin)/(1.0*ymax - ymin)
imshow(log10(hist+1),extent=[xmin,xmax,ymin,ymax],interpolation='nearest',origin='lower', aspect=aspectratio)

ax1.set_ylim([ymin,ymax])
ax1.set_xlabel("$\mathrm{x}$",fontsize=27)
ax1.set_ylabel("$\log_{10} \mathrm{ N_{scatt}}$",fontsize=27)
ax1.tick_params(axis='both', which='major', labelsize=18)

new_tick_locations = np.array([0, .166, .3333, .5, .666, .833, 1])


ax2 = ax1.twiny()
ax2.set_xticks(new_tick_locations)
ax2.tick_params(labelsize=18)
ax2.set_xticklabels(['-768','-512','-256', '0', '256', '512', '768'])
ax2.set_xlabel(r"$\mathrm{Homogeneous}$, $\mathrm{\tau=10^5}, V=0 kms^{-1}$", fontsize=27)

# Color bar
cbar = colorbar(shrink = 0.85, orientation = 'horizontal')
cbar.ax.tick_params(labelsize = 15) 
cbar.ax.xaxis.get_offset_text().set_position((1.5,1))
#savefig ('2dHistogram0t5HOM.png')
show()
```
