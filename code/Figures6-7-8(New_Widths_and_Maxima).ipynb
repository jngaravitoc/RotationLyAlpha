```python
def error(x, y, mode):
    X2 = sum(x**2)
    X = sum(x)    #X2 = (sum(xdata[i]))**2
    N = float(len(x))
    Del = N*X2 - X**2
    if mode == 0:
        sigma = sum((y - a[0] -a[1]*x)**2)
    elif mode == 1:
        sigma = sum((y - ah[0] -ah[1]*x)**2)
    sigma_y = sqrt((1/(N-2))*sigma)
    SIGMA = sigma_y*sqrt(N/float(Del))
    return  Del, sigma_y, SIGMA
l = error(xdata, ydata, 0)
lh = error(xdatah, ydatah, 1)

print l, lh
```
