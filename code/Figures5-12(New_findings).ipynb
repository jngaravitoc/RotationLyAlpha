```python
def loading_data(filename, mode):
    if mode==0:
        data = loadtxt("../data/Central/" + filename)
    elif mode==1:
        data = loadtxt("../data/Homogeneous/" + filename)
    kz = data[:,5]
    x = data[:,6]
    nscatt = data[:,8]
    return kz, x, nscatt
```

```python
#kz5t0, x5t0, nscatt = loading_data("Rotating0Sphere5t.ascii", 0)
kz5t100, x5t100, nscatt = loading_data("Rotating100Sphere5t.ascii", 0)
kz5t200, x5t200, nscatt = loading_data("Rotating200Sphere5t.ascii", 0)
kz5t300, x5t300, nscatt = loading_data("Rotating300Sphere5t.ascii", 0)

#kz5t0h, x5t0h, nscatth = loading_data("Rotating0Sphere5tHOM.ascii", 1)
kz5t100h, x5t100h, nscatth = loading_data("Rotating100Sphere5tHOM.ascii", 1)
kz5t200h, x5t200h, nscatth = loading_data("Rotating200Sphere5tHOM.ascii", 1)
kz5t300h, x5t300h, nscatth = loading_data("Rotating300Sphere5tHOM.ascii", 1)


```
