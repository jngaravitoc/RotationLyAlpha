```python
def pole_eq_histograms(data_file):
    out_data = loadtxt(data_file)
    pos_x = out_data[:,0]
    pos_y = out_data[:,1]
    pos_z = out_data[:,2]
    dir_x = out_data[:,3]
    dir_y = out_data[:,4]
    dir_z = out_data[:,5]
    out_x = out_data[:,6]
    max_x = max(pos_x)
    pos_x = pos_x/max_x
    pos_y = pos_y/max_x
    pos_z = pos_z/max_x
    pos_r = sqrt(pos_x**2 + pos_y**2 + pos_z**2)
    pos_theta = arccos(pos_z/pos_r)*180.0/pi
    pos_phi = arccos(pos_x/pos_r)*180.0/pi
    negative_y = where(pos_y<0)
    pos_phi[negative_y] = 360.0 - pos_phi
    radial_unit_x = pos_x/pos_r
    radial_unit_y = pos_y/pos_r
    radial_unit_z = pos_z/pos_r
    cos_theta_normal = (radial_unit_x * dir_x) + (radial_unit_y * dir_y) + (radial_unit_z * dir_z)
    north_pole_index = where(pos_theta<10.0)
    equator_index = where((pos_theta>82)&(pos_theta<98) & (pos_phi<90.0))
    print size(north_pole_index), size(equator_index)
    cos_theta_normal_pole = cos_theta_normal[north_pole_index]
    cos_theta_normal_equator = cos_theta_normal[equator_index]
    hist_pole, edges_pole = histogram(cos_theta_normal_pole, normed=1, bins=12)
    hist_eq, edges_eq = histogram(cos_theta_normal_equator, normed=1, bins=12)
    center_eq = 0.5*(edges_eq[1:]+edges_eq[:-1])
    center_pole = 0.5*(edges_eq[1:]+edges_eq[:-1])
    return center_eq, hist_eq, center_pole, hist_pole
```

```python
filename = "../data/Homogeneous/RotatingV0Sphere5tHOM_out.ascii"
edges_eq_v0t5, histo_eq_v0t5, edges_pole_v0t5, histo_pole_v0t5 = pole_eq_histograms(filename)
filename = "../data/Homogeneous/RotatingV300Sphere5tHOM_out.ascii"
edges_eq_v300t5, histo_eq_v300t5, edges_pole_v300t5, histo_pole_v300t5 = pole_eq_histograms(filename)
```

```python
def enlarge_labels(ax, size=20):
    ticklabels_x = ax.get_xticklabels()
    ticklabels_y = ax.get_yticklabels()
    for label_x in ticklabels_x:
        label_x.set_fontsize(size)
        label_x.set_family('serif')
    for label_y in ticklabels_y:
        label_y.set_fontsize(size)
        label_y.set_family('serif')
```

```python
fig = figure(figsize=(8.5,8.5))
ax = axes()
enlarge_labels(ax, size=30)
ax.set_xlabel("$\mu\equiv\hat{k}_{out}\cdot \hat{r}_{out}$",fontsize=30)
ax.set_ylabel("$I(\mu)$",fontsize=30)

plot(edges_eq_v0t5, histo_eq_v0t5, 
         label=r"$v=0\ $$\mathrm{km\ s^{-1}},$ $\tau=10^5$",ms=1,linewidth=3.0, color='black')
plot(edges_eq_v300t5, histo_eq_v300t5, 'r--',
     label=r"$v=300\ $ $\mathrm{km\ s^{-1}},$ $\tau=10^5$",ms=2,linewidth=3.0, color='black')

ax.legend(loc=0, scatterpoints=1, prop={'size':25}, numpoints=1)

ax.set_xlim([0.0,1.0])


filename="fig12a"
savefig(filename + '.eps',format = 'eps', transparent=True)
savefig(filename + '.pdf',format = 'pdf', transparent=True)
savefig(filename + '.png',format = 'png', transparent=True)
!pdfcrop fig12a.pdf
!mv fig12a-crop.pdf fig12a.pdf
!mv fig12a.pdf ../paper/
```
